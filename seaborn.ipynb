```python
df1=sns.load_dataset("penguins")
```

```python
df1
```

```python
sns.lineplot(x='bill_length_mm',y='flipper_length_mm',data=df1)
plt.show()
```

```python
sns.lineplot(x='bill_length_mm',y='flipper_length_mm',data=df1,hue='sex')
plt.show()
```

```python
sns.lineplot(x='bill_length_mm',y='flipper_length_mm',data=df1,hue='sex',size=10)
plt.show()
```

```python
sns.lineplot(x='bill_length_mm',y='flipper_length_mm',data=df1,hue='sex',style='sex')
plt.show()
```

```python
sns.lineplot(x='bill_length_mm',y='flipper_length_mm',data=df1,hue='sex',style='sex',palette='rainbow_r')
plt.show()
```

```python
sns.lineplot(x='bill_depth_mm',y='flipper_length_mm',data=df1,hue='sex',style='sex',palette='rainbow_r',markers=["o",">"])
plt.show()
```

```python
sns.lineplot(x='bill_depth_mm',y='flipper_length_mm',data=df1,hue='sex',style='sex',palette='rainbow_r',markers=["o",">"])
plt.grid()
plt.show()
```

------Bar plot in seaborn-----

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1
```

```python
sns.barplot(x=df1.island,y=df1.bill_length_mm)
plt.show()
```

```python
sns.barplot(x="island",y="bill_length_mm",data=df1)
plt.show()
```

```python
sns.barplot(x="island",y="bill_length_mm",data=df1,hue="sex")
plt.show()
```

```python
sns.barplot(x="island",y="bill_length_mm",data=df1,hue="sex",order=['Dream','Torgersen',"Biscoe island"])
plt.show()
```

```python
sns.barplot(x="bill_depth_mm",y="bill_length_mm",data=df1,orient='h')
plt.show()
```
